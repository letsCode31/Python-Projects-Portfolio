## Linear Relationship 
##### A linear relationship between two variables is when the change in one variable is associated with a proportional change in the other.

#### First, we'll try to plot the y=mx+c relation

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Generate synthetic data

np.random.seed(42)  # For reproducibility; the sequence of random numbers generated will always be the same each time you run the code.

X = np.random.rand(100) * 10  # Random values for X between 0 and 10
C = np.random.normal(0, 3, 100) # Noise
Y = 2.5 * X + C  # Y has a linear relationship with X plus some noise

# Step 2: Plot using Matplotlib
plt.figure(figsize=(5, 3))
plt.scatter(X, Y, color='b', alpha=0.6, label='Data points')
plt.title('Scatter Plot of X vs Y')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python

# Step 3: Plot using Seaborn with a regression line
plt.figure(figsize=(5, 3))
sns.regplot(x=X, y=Y, ci=None, scatter_kws={'alpha':0.6}, line_kws={"color": "red"})
plt.title('Linear Relationship with Regression Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

#### Let's use the instagram data set (taken from kaggle) and write the code to plot the relationships between:

##### Likes and Impressions
##### Profile Visits and Impressions
##### Profile Visits and Likes

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = 'Instagram data.csv'  # Assuming the CSV file is in the same folder as the Jupyter Notebook
data = pd.read_csv(file_path, encoding='ISO-8859-1')

# Set the style for seaborn plots
sns.set_style("whitegrid")

# Step 1: Plot the relationship between Likes and Impressions
plt.figure(figsize=(5, 3))
sns.scatterplot(x=data['Impressions'], y=data['Likes'], alpha=0.6)
plt.title('Relationship between Likes and Impressions')
plt.xlabel('Impressions')
plt.ylabel('Likes')
plt.show()

# Step 2: Plot the relationship between Profile Visits and Impressions
plt.figure(figsize=(5, 3))
sns.scatterplot(x=data['Impressions'], y=data['Profile Visits'], alpha=0.6)
plt.title('Relationship between Profile Visits and Impressions')
plt.xlabel('Impressions')
plt.ylabel('Profile Visits')
plt.show()

# Step 3: Plot the relationship between Profile Visits and Likes
plt.figure(figsize=(5, 3))
sns.scatterplot(x=data['Likes'], y=data['Profile Visits'], alpha=0.6)
plt.title('Relationship between Profile Visits and Likes')
plt.xlabel('Likes')
plt.ylabel('Profile Visits')
plt.show()
```

#### NOTE: sns.regplot(): function creates a scatter plot and fits a regression line
##### let's see how

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset  
data = pd.read_csv('Instagram data.csv', encoding='ISO-8859-1') # Ensure the CSV file is in the same folder as the Jupyter Notebook

# Set the style for seaborn plots
sns.set_style("whitegrid")

# Step 1: Plot the relationship between Likes and Impressions with a regression line
plt.figure(figsize=(5, 3))
sns.regplot(x=data['Impressions'], y=data['Likes'], scatter_kws={'alpha':0.6}, line_kws={"color": "red"})
plt.title('Relationship between Likes and Impressions')
plt.xlabel('Impressions')
plt.ylabel('Likes')
plt.show()

# Step 2: Plot the relationship between Profile Visits and Impressions with a regression line
plt.figure(figsize=(5, 3))
sns.regplot(x=data['Impressions'], y=data['Profile Visits'], scatter_kws={'alpha':0.6}, line_kws={"color": "red"})
plt.title('Relationship between Profile Visits and Impressions')
plt.xlabel('Impressions')
plt.ylabel('Profile Visits')
plt.show()

# Step 3: Plot the relationship between Profile Visits and Likes with a regression line
plt.figure(figsize=(5, 3))
sns.regplot(x=data['Likes'], y=data['Profile Visits'], scatter_kws={'alpha':0.6}, line_kws={"color": "red"})
plt.title('Relationship between Profile Visits and Likes')
plt.xlabel('Likes')
plt.ylabel('Profile Visits')
plt.show()
```
